```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

    
 #### Название колонок   
    
    Rank            Общий рейтинг
    Country         Страна или Регион
    Score           Счет
    Economy         ВВП на душу населения
    Healthy         Ожидаемая продолжительность здоровой жизни
    Generosity      Щедрость
    
    
    
  * взять только необходимые колонки
  * переименовать все колонки во всех годах для соединения
  * проставить год 

```python
df_2015 = pd.read_csv('2015.csv')

#rename collums
df_2015 = df_2015.rename(columns={'Happiness Rank': 'Rank',
                        'Happiness Score': 'Score',
                        'Economy (GDP per Capita)': 'Economy', 
                        'Health (Life Expectancy)': 'Health'})

df_2015 = df_2015[['Country', 'Rank', 'Score', 'Economy', 'Health', 'Generosity']]
df_2015['Year'] = 2015
```

```python
df_2016 = pd.read_csv('2016.csv')

#rename collums
df_2016 = df_2016.rename(columns={'Happiness Rank': 'Rank',
                        'Happiness Score': 'Score',
                        'Economy (GDP per Capita)': 'Economy', 
                        'Health (Life Expectancy)': 'Health'})

df_2016 = df_2016[['Country', 'Rank', 'Score', 'Economy', 'Health', 'Generosity']]
df_2016['Year'] = 2016
```

```python
df_2017 = pd.read_csv('2017.csv')

#rename collums
df_2017 = df_2017.rename(columns={'Happiness.Rank': 'Rank',
                        'Happiness.Score': 'Score',
                        'Economy..GDP.per.Capita.': 'Economy', 
                        'Health..Life.Expectancy.': 'Health'})

df_2017 = df_2017[['Country', 'Rank', 'Score', 'Economy', 'Health', 'Generosity']]
df_2017['Year'] = 2017
```

```python
df_2018 = pd.read_csv('2018.csv')

#rename collums
df_2018 = df_2018.rename(columns={'Overall rank': 'Rank',
                                  'Country or region': 'Country',
                                  'GDP per capita': 'Economy', 
                                  'Healthy life expectancy': 'Health'})

df_2018 = df_2018[['Country', 'Rank', 'Score', 'Economy', 'Health', 'Generosity']]
df_2018['Year'] = 2018
```

```python
df_2019 = pd.read_csv('2019.csv')

#rename collums
df_2019 = df_2019.rename(columns={'Overall rank': 'Rank',
                                  'Country or region': 'Country',
                                  'GDP per capita': 'Economy', 
                                  'Healthy life expectancy': 'Health'})

df_2019 = df_2019[['Country', 'Rank', 'Score', 'Economy', 'Health', 'Generosity']]
df_2019['Year'] = 2019
```

##### соединить dataframes

```python
df = pd.concat([df_2015, df_2016, df_2017, df_2018, df_2019], ignore_index=True)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.groupby('Year')['Score'].max().plot(title='Динамика максимальной оценки', 
                      xlabel='Год', 
                      ylabel='Максимальная оценка',
                      xticks = range(2015, 2020))
plt.legend()
plt.show()
```

```python
df.groupby('Year')['Score'].mean().plot(title='Динамика средней оценки', 
                      xlabel='Год', 
                      ylabel='Средняя оценка',
                      xticks = range(2015, 2020))
plt.legend()
plt.show()
```
