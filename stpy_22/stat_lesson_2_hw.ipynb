```python
import pandas as pd
import numpy as np
```

```python
# список для названий колонок
col_name = ['col_' + str(x) for x in range(1, 29)]
```

```python
# чтение и запись файла
df = pd.read_csv('horse_data.csv', names=col_name, na_values='?')
```

1 surgery?, операция 1да/2нет
2 Age, 1 младше6/2 старше6  
4 rectal temperature, градусы нормальная 37.8  
5 pulse,  30 -40 норм
6 respiratory rate, 7-8 норм
7 temperature of extremities, 1 норм/2теплый/3прохладный/4 холодный
11 pain, 
23 outcome. 1 = жил/2 = умер/3 = был усыплен


```python
# фильтр по нужным колонкам
df = df.filter(['col_1', 'col_2', 'col_4', 'col_5', 'col_6', 'col_7', 'col_11', 'col_23' ])
```

```python
# переименование колонк
df = df.rename(columns={'col_1': 'surgery?',
                    'col_2': 'Age',
                    'col_4': 'rectal temperature',
                    'col_5': 'pulse',
                    'col_6': 'respiratory rate',
                    'col_7': 'temperature of extremities',
                    'col_11': 'pain',
                    'col_23': 'outcome' })
```

```python
df.describe()
```

```python
# линейные показатели
#Минимум, максимум и размах
print('Минимальное значение для линейных значений')
print(df[[ 'rectal temperature','pulse', 'respiratory rate' ]].min())

print('')
print('Максимальное значение для линейных значений')
print(df[[ 'rectal temperature','pulse', 'respiratory rate' ]].max())

print('')
print('Размах')

rectal_temp_range = df['rectal temperature'].max() - df['rectal temperature'].min()
print(f'rectal_temp_range  \t {rectal_temp_range:.1f}')

pulse_range = df['pulse'].max() - df['pulse'].min()
print(f'pulse_range \t \t {pulse_range:.1f}')

respiratory_rate_range = df['respiratory rate'].max() - df['respiratory rate'].min()
print(f'resp_rate_range \t {respiratory_rate_range:.1f}')

print()
print('Среднеарифметическое')
print(df[[ 'rectal temperature','pulse', 'respiratory rate' ]].mean().round(1))

print()
print('Медиана')
print(df[[ 'rectal temperature','pulse', 'respiratory rate' ]].median().round(1))

print()
print('СКО')
print(df[[ 'rectal temperature','pulse', 'respiratory rate' ]].std().round(1))

print()
print('Дисперсия')
print(df[[ 'rectal temperature','pulse', 'respiratory rate' ]].var().round(1))

```

```python
#категорийные
#значение моды и колво вхождений

surgery_mode = df['surgery?'].round().mode()[0]
surgery_mode_count = len(df[(df['surgery?'] == (df['surgery?'].round().mode()[0])) ]['surgery?'])
print(f'Значение моды для surgery? : {surgery_mode} Кол-во вхождений: {surgery_mode_count}')

print()
age_mode = df['Age'].round().mode()[0]
age_mode_count = len(df[(df['Age'] == (df['Age'].round().mode()[0])) ]['Age'])
print(f'Значение моды для Age : {age_mode} Кол-во вхождений: {age_mode_count}')

print()
toe_mode = df['temperature of extremities'].round().mode()[0]
toe_mode_count = len(df[(df['temperature of extremities'] == (df['temperature of extremities'].round().mode()[0])) ]['temperature of extremities'])
print(f'Значение моды для temperature of extremities : {toe_mode} Кол-во вхождений: {toe_mode_count}')


print()
pain_mode = df['pain'].round().mode()[0]
pain_mode_count = len(df[(df['pain'] == (df['pain'].round().mode()[0])) ]['pain'])
print(f'Значение моды для pain : {pain_mode} Кол-во вхождений: {pain_mode_count}')

print()
outcome_mode = df['outcome'].round().mode()[0]
outcome_mode_count = len(df[(df['outcome'] == (df['outcome'].round().mode()[0])) ]['outcome'])
print(f'Значение моды для pain : {outcome_mode} Кол-во вхождений: {outcome_mode_count}')
```

# Выбросы

```python
print('Выбросы rectal temperature')
q1 = df['rectal temperature'].quantile(0.25)
q3 = df['rectal temperature'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
print(df[(df['rectal temperature'] < lower_bound) | 
         (df['rectal temperature'] > upper_bound)]['rectal temperature'])


print()
print('Выбросы pulse')
q1 = df['pulse'].quantile(0.25)
q3 = df['pulse'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
print(df[(df['pulse'] < lower_bound) | 
         (df['pulse'] > upper_bound)]['pulse'])

print()
print('Выбросы respiratory rate')
q1 = df['respiratory rate'].quantile(0.25)
q3 = df['respiratory rate'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
print(df[(df['respiratory rate'] < lower_bound) | 
         (df['respiratory rate'] > upper_bound)]['respiratory rate'])

```
