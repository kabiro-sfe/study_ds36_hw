```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('water.csv')
df = df.drop('Unnamed: 0', axis=1)
```

### Задание 1
##### Построить точечный график

```python
fig, ax = plt.subplots()
ax.scatter(df['hardness'], df['mortality'])
plt.title('Точечный график')
ax.set_xlabel('hardness')
ax.set_ylabel('mortality')
plt.show()
```

 * по точечному графику наблюдается прямая обратная линейная корреляция
 
 ##### Рассчитать коэффициенты корреляции Пирсона и Спирмена

```python
sns.heatmap(df.corr(), cmap='coolwarm',annot=True).set_title('Коэффициенты корреляции Пирсона')
plt.show()
sns.heatmap(df.corr(method='spearman'), cmap='coolwarm',annot=True).set_title('Коэффициенты корреляции Спирмана')
plt.show()
sns.heatmap(df.corr(method='kendall'), cmap='coolwarm',annot=True).set_title('Коэффициенты корреляции Кендала')
plt.show()
```

* наблюдается средний коэффициент корреляции  (< -0,63) по методу Пирсона и Спирмана
* наблюдается низкий коэффициент корреляции  ( -0,45) по методу Кендала

##### Построить модель линейной регрессии


```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
X = df[['hardness']]
y = df['mortality'] 

# X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.05, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=40)
model = LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
```

```python
fig, ax = plt.subplots()

ax.set_xlabel('hardness')
ax.set_ylabel('mortality')
ax.scatter(X, y, c='y', label = 'real')
ax.scatter(X_test, y_test , label = 'test')
ax.plot(X_test, y_pred, c='r', label = 'model')
#plt.plot(X_test, y_pred, c='r')
ax.legend(loc='best')
plt.title('Модель линейной регрессии')
plt.show()
```

```python
print('Модель линейной регрессии Seaborn')
sns.regplot(x='hardness', y='mortality', data=df)
plt.title('Модель линейной регрессии v.2')
plt.show()
```

##### Рассчитать коэффициент детерминации

```python
print(round(model.score(X_test, y_test), 4), '- средний коэффициент детерминации') 
```

##### Вывести график остатков

```python
res_df = X_test.copy()
res_df['y_pred'] = model.predict(X_test)
res_df['y_test'] = y_test
res_df['res'] = res_df['y_pred'] - res_df['y_test']
```

```python
fig, ax = plt.subplots()
ax.scatter(res_df['hardness'], res_df['res'])
plt.title('График остатков')
ax.set_xlabel('hardness')
ax.set_ylabel('res')
plt.show()
```

```python
fig, ax = plt.subplots()
ax.hist(res_df['res'])
plt.title('График вхождений остатков')
ax.set_xlabel('res')
ax.set_ylabel('count')
plt.show()
```

Остатки ведут себя разрознено. закономерностей не наблюдается. Можно предположить, что модель построена достаточно верно

#### Ответить на вопрос: есть ли связь между жёсткостью воды и средней годовой смертностью?
Математическая модель показывает, что некоторая связь между смертностью и жесткостью воды имеется. Возможно при наличии еще данных это может быть и подтверждено и опровергнуто.
