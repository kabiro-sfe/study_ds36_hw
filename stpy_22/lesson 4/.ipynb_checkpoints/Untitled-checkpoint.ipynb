```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Полные данные

```python
df = pd.read_csv('water.csv')
df = df.drop(columns=['Unnamed: 0'])
```

```python
df.plot(kind='scatter', y='mortality', x='hardness').set_title('Влияние жесткости воды на смертность')
plt.show()

print('есть визуальная зависимость')
```

```python
sns.heatmap(df.corr(), cmap='coolwarm',annot=True).set_title("коэффициента Пирсона", )
plt.show()
sns.heatmap(df.corr(method='spearman'), cmap='coolwarm',annot=True).set_title('метод Спирмана')
plt.show()
sns.heatmap(df.corr(method='kendall'), cmap='coolwarm',annot=True).set_title('метод Кендала')
plt.show()
```

высокая обратная зависимоть по методу Пирсона и Спирмана

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
X = df[['hardness']]
y = df['mortality']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)

model = LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
```

```python
model.score(X_test, y_test) # низкий коеф детерминации
```

```python
plt.scatter(X, y, c='y', label = 'real')
plt.scatter(X_test, y_test , label = 'test')
plt.plot(X_test, y_pred, c='r', label = 'model')
#plt.plot(X_test, y_pred, c='r')
plt.legend(loc='best')
plt.title('Проверка модели Линейная')
plt.show()
```

```python
sns.regplot(x='hardness', y='mortality', data=df)
plt.title('Линия регрессии')
plt.show()
```
