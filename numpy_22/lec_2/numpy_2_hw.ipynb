```python
data = pd.read_csv('ml-25m/ratings.csv')

res = data[(data['rating']==5)].value_counts(['movieId']).head(1).to_frame()


mov = pd.read_csv('ml-25m/movies.csv')

result = res.merge(mov, on='movieId', how='left')

result.filter(['movieId', 'title'])
```

### Задание 2

```python
data = pd.read_csv('10/power.csv')


data_fil = (data[((data['country'] == 'Latvia') |  
                  (data['country'] == 'Lithuania') |  
                  (data['country'] == 'Estonia') ) & # отбор по странам
                 ((data['year'] < 2010) & (data['year'] > 2005)) &  # отбор по категориям
                    ((data['category'] == 4) | (data['category'] == 12) | (data['category'] == 21)) & # отбор по категориям
                    (data['quantity'] > 0)]) # без отрицательных значений

res = data_fil.groupby(['country']).sum().filter(['country', 'quantity'])
res
```
